```python

# Display first few rows
print("Original Dataset:")
print(df.head())

# for selection of numerical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Loop through each numerical column and apply IQR
for col in numeric_cols:
    
    # Calculate Q1 and Q3
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nColumn: {col}")
    print("Lower Bound:", lower_bound)
    print("Upper Bound:", upper_bound)
    print("Number of Outliers:", outliers.shape[0])

# Remove outliers using IQR
for col in numeric_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("\nDataset after removing outliers:")
print(df.shape)

# Save cleaned dataset
#df.to_csv("cleaned_data.csv", index=False)
```

```python
#create two new feature hydraulic power and efficiency for better data prediction
df['hydraulic_power'] = (df['pressure'] * df['flow_rate']) / 1714
print(df.head())
```

```python
df['Efficency'] = (df['hydraulic_power'] / df['power_consumption'])*100
print(df.head())
```

```python
#saved all the data processing in seperate csv file to avoid any change in default csv file
df.to_csv("cleaned_data.csv", index=False)
```
